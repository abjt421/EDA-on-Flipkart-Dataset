# **Exploratory Data Analysis on Pricing of Flipkart Dataset**

The dataset comprises information about various products available on Flipkart after undergoing preprocessing, resulting in a refined dataset containing **19,438 entries** over **10 columns**. The selected features include:

- **Brand**: The brand name of the product.
- **Product Name**: The name or title of the product.
- **Month**: The month of the observation, indicating when the product data wasolded.
- **Category**: The category to which the product belongs, providing insights into the product type or classification.
- **Retail Price**: The original price of the product before any discount.
- **Discounted Price**: The price of the product after applying discounts.
- **Discount Percent**: The percentage of discount applied to the product.
- **is_FK_Advantage_Product**: A binary indicator representing whether the product is part Flipkartlatform's "FK Advantage" program, which could offer benefits or advantages to customers.
- **Product Rating**: The rating given to the product by customers.
- **Overall Rating**: The overall rating, potentially aggregated from various aspects or criteria.


## **Loading the Dataset**

```python
import pandas as pd
df = pd.read_csv("final_flipkart.csv")
df.sample(10)
```

## **Pricing based analysis**

### Clustered Column Graph for Retail and Discounted Prices

```python
import matplotlib.pyplot as plt

bins_retail = [i for i in range(0, 60001, 5000)]
df['price_range_retail'] = pd.cut(df['retail_price'], bins=bins_retail, right=False)
price_counts_retail1 = df['price_range_retail'].value_counts().sort_index()

bins_discounted = [i for i in range(0, 45001, 5000)]
df['price_range_discounted'] = pd.cut(df['discounted_price'], bins=bins_discounted, right=False)
price_counts_discounted1 = df['price_range_discounted'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(range(len(price_counts_retail1)), price_counts_retail1, color='blue', label='Retail Price', width=0.4)
plt.bar([x + 0.4 for x in range(len(price_counts_discounted1))], price_counts_discounted1, color='red', width=0.4, label='Discounted Price')

for i, val in enumerate(price_counts_retail1):
    plt.text(i, val + 1, str(val), ha='center', va='bottom', fontsize=9)
for i, val in enumerate(price_counts_discounted1):
    plt.text(i + 0.4, val + 1, str(val), ha='center', va='bottom', fontsize=9)

plt.xticks([x + 0.2 for x in range(len(price_counts_retail1))], ['0 - 5000', '5000 - 10000', '10000 - 15000', '15000 - 20000', '20000 - 25000', '25000 - 30000', '30000 - 35000', '35000 - 40000', '40000 - 45000', '45000 - 50000', '50000 - 55000', '55000 - 60000'], rotation=45)

plt.title('Bar Graph of Retail and Discounted Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.grid(axis='y')
plt.legend()

plt.tight_layout()
plt.show()
```

**Maximum products were sold in the price range of 0 to 5000**

## **What is mean, median and mode for Retail Price, Discounted Price and Discount Percent respectively ??**

```python
mean_retail_price = df['retail_price'].mean()
median_retail_price = df['retail_price'].median()
mode_retail_price = df['retail_price'].mode()

mean_discounted_price = df['discounted_price'].mean()
median_discounted_price = df['discounted_price'].median()
mode_discounted_price = df['discounted_price'].mode()

mean_discount_percent = df['discount_percent'].mean()
median_discount_percent = df['discount_percent'].median()
mode_discount_percent = df['discount_percent'].mode()

mean_df = pd.DataFrame({'Metric': 'Mean', 'Retail Price': mean_retail_price, 'Discounted Price': mean_discounted_price, 'Discount Percent': mean_discount_percent}, index=[0])
median_df = pd.DataFrame({'Metric': 'Median', 'Retail Price': median_retail_price, 'Discounted Price': median_discounted_price, 'Discount Percent': median_discount_percent}, index=[0])
mode_df = pd.DataFrame({'Metric': 'Mode', 'Retail Price': mode_retail_price.values, 'Discounted Price': mode_discounted_price.values, 'Discount Percent': mode_discount_percent.values})
summary_df = pd.concat([mean_df, median_df, mode_df], ignore_index=True)
summary_df
```

## **Discount based analysis**

```python
intervals = [i * 10 for i in range(11)]
df['discount_intervals'] = pd.cut(df['discount_percent'], bins=intervals, right=False)
interval_counts = df['discount_intervals'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
ax = interval_counts.plot(kind='bar', color='darkblue')

for i, val in enumerate(interval_counts):
    ax.text(i, val + 1, str(val), ha='center', va='bottom', fontsize=9)

ticks = range(len(interval_counts))
labels = [f'{i} - {i + 10}' for i in intervals[:-1]]
plt.xticks(ticks, labels, rotation=45)

plt.title('Bar Graph of Discount Percent Intervals')
plt.xlabel('Discount Percent Intervals')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### **50 - 60 % discount was provided more frequently.**

```python
import pandas as pd
import matplotlib.pyplot as plt

intervals = [i * 10 for i in range(11)]
df['discount_intervals'] = pd.cut(df['discount_percent'], bins=intervals, right=False)
interval_counts = df['discount_intervals'].value_counts().sort_index()

plt.figure(figsize=(6, 6))
colors = plt.cm.Dark2.colors[:len(interval_counts)]
plt.pie(interval_counts, labels=None, colors=colors, autopct='%1.1f%%', startangle=0)

legend_labels = [f'{interval.left} - {interval.right} ({pc:.1f}%)' for interval, pc in zip(interval_counts.index, interval_counts / interval_counts.sum() * 100)]
plt.legend(labels=legend_labels, loc='upper right', bbox_to_anchor=(1.2, 1))

plt.title('Pie Chart of Discount Percent Intervals')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## **Month based Analysis**

```python
month_counts = df['month'].value_counts()

plt.figure(figsize=(8, 8))
colors = plt.cm.tab20.colors[:len(month_counts)]
patches, texts, _ = plt.pie(month_counts, colors=colors, autopct='%1.1f%%', startangle=90)

legend_labels = [f'{month} ({count}, {percent:.1f}%)' for month, count, percent in zip(month_counts.index, month_counts, month_counts / month_counts.sum() * 100)]
plt.legend(patches, legend_labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Pie Chart of Month Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

**More than 50% products were sold in December 2015 only.**

```python
df['month'] = pd.to_datetime(df['month'])
df_sorted = df.sort_values('month')

median_discounted_price_by_month = df_sorted.groupby('month')['discounted_price'].median()

plt.figure(figsize=(10, 6))
median_discounted_price_by_month.plot(kind='bar', color='skyblue')

plt.title('Median Final Price of Products Sold in Different Months')
plt.xlabel('Month')
plt.ylabel('Median Final Price')
plt.xticks(rotation=0)
plt.gca().set_xticklabels([x.strftime('%B %Y') for x in median_discounted_price_by_month.index])
plt.tight_layout()
plt.show()
```

**But, products worth more final price was sold in January 2016.**

## **Category based Analysis**

**We are considering only those categories which have more than 100 products.**
**Final price is the discounted price.**

```python
category_counts = df['category'].value_counts()
filtered_category_counts = category_counts[category_counts > 100]
filtered_category_counts
```

```python
category_counts = df['category'].value_counts()
filtered_category_counts = category_counts[category_counts > 100]
plt.figure(figsize=(12, 6))
bars = plt.bar(filtered_category_counts.index, filtered_category_counts.values, color='lightcoral')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, yval, ha='center', va='bottom', fontsize=8)

plt.title('Value Counts for Categories')
plt.xlabel('Category')
plt.ylabel('Value Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
total_products = len(df)
percent_of_total = (filtered_category_counts / total_products) * 100

plt.figure(figsize=(12, 6))
bars = plt.bar(filtered_category_counts.index, percent_of_total, color='green')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.1f}%', ha='center', va='bottom', fontsize=10)

plt.title('Percentage of Total Products for Categories')
plt.xlabel('Category')
plt.ylabel('Percentage of Total Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### **About one - third of the products sold are from Clothing category.**

```python
df_filtered = df[df['category'].isin(filtered_category_counts.index)]

median_discounted_price_by_category = df_filtered.groupby('category')['discounted_price'].median()

plt.figure(figsize=(12, 6))
bars = plt.bar(median_discounted_price_by_category.index, median_discounted_price_by_category.values, color='grey')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, f'{yval:.2f}', ha='center', va='bottom', fontsize=8)

plt.title('Median Final Price for Categories with Counts > 100')
plt.xlabel('Category')
plt.ylabel('Median Final Price')
plt.xticks(rotation=45, ha='right')
plt.grid(axis = "y")
plt.tight_layout()
plt.show()
```

## **Rating based Analysis**

```python
df['rating_status'] = df['overall_rating'].notna()
rating_counts = df['rating_status'].value_counts()

plt.figure(figsize=(6, 6))
colors = plt.cm.tab20.colors[:len(rating_counts)]
patches = plt.pie(rating_counts, labels=None, colors=colors, autopct='%1.1f%%', startangle=90)

legend_labels = ['Rated' if idx else 'Not Rated' for idx in rating_counts.index]

plt.legend(patches[0], legend_labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Pie Chart of Rating Status')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### **Only 9% products were rated.**

### **Pie Chart of Ratings with non - Nan Values**

```python
rating_ranges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 5)]

rating_counts = [((df['overall_rating'] >= low) & (df['overall_rating'] < high)).sum() for low, high in rating_ranges]

labels = ['{}-{}'.format(low, high) for low, high in rating_ranges]
plt.pie(rating_counts, labels=labels, autopct='%1.1f%%')
plt.title('Distribution of Overall Ratings')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.legend(labels, loc='upper right', bbox_to_anchor=(1.2, 1))

plt.show()
```
